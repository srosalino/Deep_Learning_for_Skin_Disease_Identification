This Notebook was used to download the images using the provided URL in the original dataset into a local folder (images). Additionally, a column was added to the resulting processed dataset containing the path to the corresponding image in the local folder.

```python
# Import the necessary libraries and configurations
import pandas as pd
import requests
from PIL import Image
from io import BytesIO
import os
```

```python
# Reading the original dataset
df = pd.read_csv('fitzpatrick17k.csv')
```

```python
df.head()
```

# Download Images and Write Path

```python
def download_and_save_image(url, filepath, target_dir, image_id, total_images):
    try:
        
        # Make sure the target directory exists
        os.makedirs(target_dir, exist_ok=True)
        
        # Download the image
        headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.3'}
        response = requests.get(url, headers=headers)
        response.raise_for_status()  # This will raise an error for bad responses
        
        # Open the image from bytes and convert it to RGB
        image = Image.open(BytesIO(response.content)).convert('RGB')
        
        # Save the image
        image.save(filepath)
        
        # Return the path where the image was saved
        print(f"Processed {image_id+1}/{total_images} images. URL: {url}")
        
        return filepath
        
    except Exception as e:
        print(f"An error occurred while processing image {image_id+1} (URL: {url}): {e}")
        return pd.NA
```

```python
total_images = len(df)
target_dir = 'images'

df['image_path'] = [
    download_and_save_image(
        url=url,
        filepath = os.path.join(target_dir, df.at[i, 'url_alphanum']),
        target_dir=target_dir,
        image_id=i,
        total_images=total_images
    )
    for i, url in enumerate(df['url'])
]
```

```python
# Exporting the resulting processed dataset
df.to_csv('fitzpatrick17k_processed.csv', index=False)
```
